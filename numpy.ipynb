```python

# From Python Lists
import numpy as np
arr = np.array([1, 2, 3])
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print(arr)
print(arr_2d)
```

```python
# Special Arrays
print(np.zeros((3, 3)))
print(np.ones((2, 4)))
print(np.eye(3))
print(np.arange(0, 10, 2))
print(np.linspace(0, 1, 5))
print(np.random.rand(3, 2))
print(np.random.randn(3, 2))
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr.shape)
print(arr.ndim)
print(arr.size)
print(arr.dtype)
```

```python
# Basic Math
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a+b)
print(a*b)
print(np.sin(a))
```

```python
# Matrix Multiplication
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(np.dot(A, B))
print(A @ B)
```

```python
# Aggregation
arr = np.array([1, 2, 3, 4])
print(arr.sum())
print(arr.mean())
print(arr.max())
print(arr.min())
print(arr.std())
```

```python
# 1D Arrays
arr = np.array([1, 2, 3, 4, 5])
print(arr[0])
print(arr[1:4])
print(arr[::2])
```

```python
# 2D Arrays
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr[0, 1])
print(arr[:, 1])
print(arr[1, :2])
```

```python
# Boolean Indexing
arr = np.array([1, 2, 3, 4, 5])
mask = arr > 3
print(arr[mask])
```

```python
# Reshape
arr = np.arange(6)
print(arr.reshape(2, 3))
```

```python
# Concatenation
a = np.array([1, 2])
b = np.array([3, 4])
print(np.concatenate((a, b)))
```

```python
# Stacking
print(np.vstack((a, b)))
print(np.hstack((a, b)))
```

```python
#broadcasting
a = np.array([1, 2, 3])
a + 5  
```
